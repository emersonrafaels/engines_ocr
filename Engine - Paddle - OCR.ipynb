# IMPORTING LIBRARIES

```python
from paddleocr import PaddleOCR, draw_ocr
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

## READING IMAGE

```python
# IMAGEM A SER ANALISADA
img_path = '/content/drive/MyDrive/DATASETS/LARGE RECEIPT/1029-receipt.jpg'
```

```python
img = cv2.imread(img_path)
```

```python
cv2_imshow(img)
```

## EXECUTE PADDLE

### ENGLISH

```python
# INSTANCIANDO O MODELO

ocr = PaddleOCR(use_angle_cls=True, lang='en') # need to run only once to download and load model into memory
```

```python
# APLICANDO O OCR NA IMAGEM
result = ocr.ocr(img, cls=True)
```

```python
# OBTENDO O RESULTADO DO OCR
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)
```

```python
result_filter = result[0]
boxes = [line[0] for line in result_filter]
txts = [str(line[1][0]) for line in result_filter]
scores = [line[1][1] for line in result_filter]
im_show = draw_ocr(img, boxes, txts, 
                   scores, 
                   font_path="/content/drive/MyDrive/arial_narrow_7.ttf")
cv2_imshow(im_show)
```

```python
list_result = []

for line_num, detection in enumerate(result):
  for value_detection in detection:
    # OBTENDO O TEXTO
    text = value_detection[-1][0]

    # OBTENDO A CONFIANÇA
    conf = value_detection[-1][1]

    # OBTENDO OS PONTOS DAS COORDENADAS
    p1 = value_detection[0][0]
    p2 = value_detection[0][1]
    p3 = value_detection[0][2]
    p4 = value_detection[0][3]
    left = p1[0]
    width = p1[1]
    height = p3[0] - p1[0]
    top = p3[1] - p1[1]

    list_result.append({"line_num": line_num, 
                        "left": left, 
                        "top": top, 
                        "width": width, 
                        "height": height, 
                        "conf": conf, 
                        "text": text})
```

# EAST

```python
import cv2
from paddleocr import PaddleOCR, draw_ocr
from paddleocr import text_detection
```

```python
# Load image
img = cv2.imread('example_image.png')

# Initialize PaddleOCR
ocr = PaddleOCR(lang='en')

# Perform text detection using EAST
east = text_detection.EASTDetector()
east_result = east.detect(img)

# Combine the results from EAST and PaddleOCR
ocr_result = ocr.ocr(img)

# Draw bounding boxes around detected text
for box in east_result:
    cv2.rectangle(img, (int(box[0]), 
                        int(box[1])), 
                  (int(box[2]), 
                   int(box[3])), 
                  (0, 0, 255), 
                  2)

for line in ocr_result:
    for word in line:
        cv2.rectangle(img, (int(word[0][0]), int(word[0][1])), (int(word[0][2]), int(word[0][3])), (0, 255, 0), 2)

# Display the image
cv2.imshow('Detected Text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
