```python
from paddleocr import PaddleOCR, draw_ocr
```

```python
# INSTANCIANDO O MODELO

ocr = PaddleOCR(use_angle_cls=True, lang='en') # need to run only once to download and load model into memory
```

```python
# IMAGEM A SER ANALISADA
img_path = '/content/IMAGENS/1029-receipt.jpg'
```

```python
# APLICANDO O OCR NA IMAGEM
result = ocr.ocr(img_path, cls=True)
```

```python
# OBTENDO O RESULTADO DO OCR
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)
```

```python
import pandas as pd
```

```python
list_result = []

for line_num, detection in enumerate(result):
  for value_detection in detection:
    # OBTENDO O TEXTO
    text = value_detection[-1][0]

    # OBTENDO A CONFIANÇA
    conf = value_detection[-1][1]

    # OBTENDO OS PONTOS DAS COORDENADAS
    p1 = value_detection[0][0]
    p2 = value_detection[0][1]
    p3 = value_detection[0][2]
    p4 = value_detection[0][3]
    left = p1[0]
    width = p1[1]
    height = p3[0] - p1[0]
    top = p3[1] - p1[1]

    list_result.append({"line_num": line_num, 
                        "left": left, 
                        "top": top, 
                        "width": width, 
                        "height": height, 
                        "conf": conf, 
                        "text": text})
```

```python
# DICT RESULTADO

dict_result = list_result.copy()
dict_result
```

```python
# DATAFRAME RESULTADO

df_result = pd.DataFrame(list_result)
df_result
```
